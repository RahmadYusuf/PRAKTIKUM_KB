## 5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.

Disini saya memilih grafik
 data menggunakan pie chart yg mana menggunakan library matplotlib setiap potongan menunjukkan satu kategori lengkap dengan persentasenya, sehingga kita bisa langsung melihat perbandingan jumlah responden pada tiap jenjang akademik secara visual.

```python
import matplotlib.pyplot as plt
data = df['Your Academic Stage'].value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Proporsi Responden per Jenjang Akademik')
plt.show()
```

## 6. Tampilkan korelasi dataset menggunakan heatmap.

Sebelum membuat heat map kita harus menggunakan library seaborn, Pada slide kedua kode akan menghitung nilai korelasi hanya pada kolom bertipe numerik, karena korelasi pada heatmap lebih bagus menggunakan data angka Nanti Hasilnya divisualisasikan dengan sns.heatmap.

```python
import seaborn as sns
cor_matrix = df.corr(numeric_only=True)
sns.heatmap(cor_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Heatmap Korelasi Atribut Numerik")
plt.show()
```
